```python
import pandas as pd

df = pd.read_csv('./partidos.csv', sep=';')
```

```python
df['Fecha_Completa'] = pd.to_datetime(df['Fecha'] + ' ' + df['Hora'], dayfirst=True)

df_ordenado = df.sort_values(by='Fecha_Completa').reset_index(drop=True)

diccionario_partidos = {i + 1: row['Numero'] for i, row in df_ordenado.iterrows()}

print(diccionario_partidos)
```
